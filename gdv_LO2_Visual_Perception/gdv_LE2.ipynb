```python
# Module importiert
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Daten laden
data = pd.read_csv("pokemon_data.csv")
```

```python
# Daten filtern (nur Feuer Pokemon)
data = data[data["Type 1"] == "Fire"]
data.info()
```

```python
# Attack/Defense und Sp. Attack/Sp. Defense Verhaeltnis erstellen
data["Atk/Def"] = (data["Attack"] + data["Sp. Atk"]) / (data["Defense"] + data["Sp. Def"])
data["Total_stats"] = data["HP"] + data["Attack"] + data["Sp. Atk"] + data["Defense"] + data["Sp. Def"] + data["Speed"]
data = data.sort_values("Total_stats", ascending=False)
# Bei Atk/Def, Wenn < 1, Defense hoeher als Attack (Je kleiner der Wert, desto hoeher der Unterschied)
# Bei Atk/Def, wenn = 1, Defense = Attack
# Bei Atk/Def, wenn > 1, Defense tiefer als Attack (Je groesser der Wert, desto hoeher der Unterschied)
```

```python
# Feuerpokemon mit hohen Angriffswerten und hohen Total Stats
data.sort_values(["Atk/Def","Total_stats"], ascending=[False,False]).head(3)
```

```python
# Feuerpokemon mit hohen Verteidigungswerten und hohen Total Stats
data.sort_values(["Atk/Def","Total_stats"], ascending=[True,False]).head(3)
```

```python
# Data Visualisation  
data.plot.scatter("Total_stats", "Atk/Def",
                  figsize=(20,10),
                  color = "grey", 
                  edgecolor = "black", 
                  linewidth = 1, 
                  s = 300, 
                  marker = "o", 
                  alpha = 1)
plt.title("Fire Pokemon Stats", fontsize = 30)
plt.xlabel("Total Stats", fontsize = 20)
plt.ylabel("Total Attack / Total Defense", fontsize = 20)
plt.style.use("seaborn-whitegrid")
plt.show()
```
