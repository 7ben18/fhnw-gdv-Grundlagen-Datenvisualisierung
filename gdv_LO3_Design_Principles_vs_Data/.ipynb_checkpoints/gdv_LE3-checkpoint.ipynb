```python
# Module importieren
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Daten in pd einlesen
temperatures = pd.read_csv("temperatures.csv", index_col="Unnamed: 0")

# Datentyp ausgeben
print(temperatures.dtypes)

# Datentyp umwandeln
temperatures["date"] = pd.to_datetime(temperatures["date"])
print(temperatures.dtypes)
```

```python
# Selektieren von 5 Laender (Chile, Vietnam, Ukraine, Sudan, Australia)
liste = ["Chile", "Vietnam", "Ukraine", "Sudan", "Australia"]
temperatures = temperatures[temperatures["country"].isin(liste)]
print(temperatures)

# Jahres Spalte hinzufuegen
temperatures["year"] = temperatures["date"].dt.year
print(temperatures) 
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x = temperatures["year"], y = temperatures["avg_temp_c"], hue = temperatures["country"])
#plt.title("Durchschnittliche Temperatur im laufe der Zeit")
plt.xlabel("Jahr")
plt.show()
```

```python
# Gruppieren nach Land
temp_group = temperatures.groupby(["country","year"])["avg_temp_c"].mean()
temp_group
```

```python
# Pivotieren der Tabellen
temp_pivot = temperatures.pivot_table(values="avg_temp_c", index="country", columns="year", aggfunc=np.mean)
temp_pivot

# selektieren von 5 Laender (Chile, Vietnam, Ukraine, Sudan, Australia)
```

```python

ax = temp_pivot.T.plot(kind = "line", 
                      figsize=(20,10),
                      ylabel="Average Temperature over the years",
                      title="Average Temperature in °C")

ax.legend(title = "artist name")
plt.show()
```
