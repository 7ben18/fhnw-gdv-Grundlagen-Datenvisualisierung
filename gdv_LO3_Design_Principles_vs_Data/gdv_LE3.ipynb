```python
# Module importieren

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Daten in pd einlesen
temperatures = pd.read_csv("temperatures.csv", index_col="Unnamed: 0")
temperatures
```

```python
# Datentyp umwandeln
temperatures["date"] = pd.to_datetime(temperatures["date"])

# Jahres Spalte hinzufuegen
temperatures["year"] = temperatures["date"].dt.year
temperatures
```

```python
# Country und City als Index setzen
temperatures = temperatures.set_index(["country", "city"])

# Index sortieren
temperatures.sort_index(level = ["country", "city"], ascending = [True, True])
```

```python
# Pivotieren 
temp_by_country_city_vs_year = temperatures.pivot_table("avg_temp_c", index=["country","city"],
                                                       columns="year", aggfunc=np.mean)  
temp_by_country_city_vs_year
```
